```python
from scipy import signal
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
sgi = np.repeat([0., 1., 1., 0., 1., 0., 0., 1.], 128)
```

```python
sig_noise = sgi + np.random.randn(len(sgi))
```

```python
plt.plot(np.arange(len(sgi)), sig_noise)
plt.show()
```

```python
corrs = signal.correlate(np.ones(len(sgi)), sig_noise, mode='same') / 128
```

```python
corr = signal.correlate(sig_noise, np.ones(len(sgi)), mode='same') / 128
```

```python
plt.plot(np.arange(len(corr)), corr)
plt.show()
```

```python
plt.plot(np.arange(len(corrs)), corrs)
plt.show()
```
